Diketahui sebuah kumpulan data sebagai berikut:

Anggaplah data berikut dari big data csv, dan Anda diminta untuk membersihkan data dari inconsistent value.

```python
import pandas as pd
data= {'person': ['Homer', 'Marge', 'Bart', 'Lisa', 'Maggie', 'Abe', 'Selma', 'Otto', 'Krusty'],
'hair_length': [ '0"', '10"', '2', '6"', '4"', '1"', '8"', '10', '6"'],
'weight':[250,150,90,78,20,170,160,180,200],
'age': [36,34,10,8,1,70,41,38,45],
'class': ['M', 'F', 'M', 'F', 'F', 'M', 'F', 'M', 'M']}
df = pd.DataFrame(data)
```

```python
df
```

Pertanyaan:

1. Ubahlah nilai atribut 'Hair Length' yang masih terdapat satuan inch (“) ke dalam data numerik.

2. buatlah kategori hair length sebagai berikut:

 0-5 = pendek 

 6-10 = panjang

1. Mengubah ke dalam numerik

```python
# mengubah satuan inch menjadi numerik
df['hair_length'] = df['hair_length'].str.replace('"', '').astype(int)
```

```python
df
```

2. Membuat kategori

```python
# Membuat kategori hair length
def hair_length_category(row):
    if row['hair_length'] >= 0 and row['hair_length'] <= 5:
        return 'pendek'
    elif row['hair_length'] >= 6 and row['hair_length'] <= 10:
        return 'panjang'
```

```python
# Menambahkan kolom baru 'hair_length_category'
df['hair_length_category'] = df.apply(hair_length_category, axis=1)
```

# **Latihan (Statistika Deskriptif)**

Untuk lebih memahami materi statistika deskriptif, silakan untuk mencoba latihan ini

Untuk periode 2001–2008, Bristol-Myers Squibb
Company, Inc. melaporkan jumlah (dalam milyaran
dolar) untuk (1) penjualan bersih dan (2) iklan dan
promosi produk \\

| Tahun | Net sales | Iklan |
|----|---|----|
|2001 | 16.612 |1.201|
|2002 |16.208 |1.143|
|2003 |18.653 |1.416|
|2004 |19.380 |1.411|
|2005 |19.207 |1.476|
|2006 |16.208 |1.304|
|2007 |18.193 |1.415|
|2008 |20.597| 1.550|

(a) Tunjukkan berapa rata-rata net sales dan iklan selama 9 tahun? \\
(b) Berapa variasi yang didapatkan untuk net sales dan iklan selama 9 tahun? \\


```python
# Membuat series dictionary
daftar = {'Tahun' : pd.Series([2001,2002,2003,2004,2005,2006,2007,2008]),
   'Net Sales' : pd.Series([16.612,16.208,18.653,19.380,19.207,16.208,18.193,20.597]),
   'Iklan' : pd.Series([1.201,1.143,1.416,1.411,1.476,1.304,1.415,1.550])}

# Display Dataframe
df = pd.DataFrame(daftar)
df
```

a. rata-rata net sales dan iklan

```python
# Mendapatkan rata-rata secara manual
# Mendapatkan mean pada kolom usia
mean_netsales = sum(df['Net Sales'])/len(df['Net Sales']) #

# Mendapatkan mean pada kolom rating
mean_iklan = sum(df['Iklan'])/len(df['Iklan'])

# Display hasil
print("Nilai rata-rata Net Sales adalah " + str(mean_netsales))
print("Nilai rata-rata Iklan adalah " + str(mean_iklan))
```

b. variasi

```python
df.var()
```

# Latihan (Statistika Deskriptif di bisnis)

Berdasarkan data transaksi dari coffee shop diatas. Carilah persebaran umur berdasarkan mean, dan varian dari customer tersebut. Serta dapatkan modus usia dari customers.

```python
df_cust = pd.read_csv('/content/drive/MyDrive/SIB/2. Statistics for Data Science/customer.csv')
```

- persebaran umur berdasarkan mean dan varian customer

```python
# Menampilkan 5 data teratas dari dataset customers
df_cust.head()
```

```python
import datetime as dt
dt.datetime.now()
df_cust['birthdate'] = pd.to_datetime(df_cust['birthdate'])
df_cust['age'] = (dt.datetime.now() - df_cust['birthdate']).dt.days/365
df_cust
```

```python
#mean dari umur
mean_umur = df_cust['age'].mean()
print('mean dari umur customer adalah ', mean_umur)

#varian dari umur
var_umur = df_cust['age'].var()
print('mean dari umur customer adalah ', var_umur)

#modus dari umur
modus_umur = df_cust['age'].mode()
print('mean dari umur customer adalah ', modus_umur)
```

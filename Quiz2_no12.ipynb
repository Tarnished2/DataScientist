Sebuah perusahaan melakukan perubahan pada layout dari produk-produk yang dimiliki. Perusahaan ingin menguji apakah layout yang baru lebih sukai oleh customer, diperoleh data sebagai berikut.

Outcome             Initial Layout           New Layout

Conversion          239                      318

Non Conversion      82700                    78200

 

Bagaimana kesimpulan yang diambil oleh perusahaan? Lakukan perhitungan manual untuk pengambilan keputusan.

1. Perhitungan manual

Conversion rate layout awal : 239 / (239 + 82700) = 0.29%

Conversion rate layout baru : 318 / (318 + 78200) = 0.41%

Selanjutnya, kita bisa menghitung nilai relative uplift di antara conversion rates: (0.41%-0.29%)/0.29% = 0.413

Menetapkan standar nilai uplift, yang biasa digunakan adalah 0,06

Kesimpulan : Nilai relative uplift lebih besar dari 0,06 maka kita bisa menyimpulkan, new layout lebih menarik bagi customer.

```python
from scipy.stats import chi2_contingency

# Input data
observed = [[239, 318], [82700, 78200]]

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(observed)

# Set significance level
alpha = 0.05

# Print results
if p < alpha:
    print("Reject null hypothesis. New layout is preferred by customers.")
else:
    print("Do not reject null hypothesis. No significant preference for new layout.")
```

berdasarkan chi-test, maka hasil dari chi-test sesuai dengan perhitungan manual.sehingga didapat kesimpulan bahwa new layout lebih menarik bagi customer.

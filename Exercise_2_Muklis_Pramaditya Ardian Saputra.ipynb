```python
df_filtered1 = df[(df['sepal.length'] >= 4) & (df['sepal.length'] <= 5)]
print(df_filtered1)
```

- filter untuk mendapatkan Sepal.Width antara 2-3

```python
df_filtered2 = df[(df['sepal.width'] >= 2) & (df['sepal.width'] <= 3)]
print(df_filtered2)
```

sepal.width

petal.length

```python
count_species = df.groupby('petal.length').size()
print(count_species)
```

petal.width

```python
count_species = df.groupby('petal.width').size()
print(count_species)
```

variety

```python
count_species = df.groupby('variety').size()
print(count_species)
```

```python
df_setosa = df[(df['sepal.width'] == 0) & (df['sepal.length'] == 0) & (df['petal.width'] == 0) & (df['petal.length'] == 0) & (df['variety'] == 0)]
print(df_setosa)
```

<h4>Lakukan latihan berikut<h4>
Diberikan angka random sebagai berikut (0.26, 0.96, 0.63, 0.5 , 0.31, 0.94, 0.71, 0.81, 0.11). Simpanlah kedalam array1 yang merupakan numpy array berdimensi 3x3 dengan menggunakan nested loop for dan while (sehingga ada 2 jawaban) ! Urutan angka bebas

- For loop

```python
import numpy as np

random_nums = [0.26, 0.96, 0.63, 0.5, 0.31, 0.94, 0.71, 0.81, 0.11]
array1 = np.zeros((3, 3))

for i in range(3):
    for j in range(3):
        array1[i][j] = random_nums[i*3+j]
        
print(array1)
```

- While loop

```python
import numpy as np

random_nums = [0.26, 0.96, 0.63, 0.5, 0.31, 0.94, 0.71, 0.81, 0.11]
array1 = np.zeros((3, 3))

i, j = 0, 0
while i < 3:
    while j < 3:
        array1[i][j] = random_nums[i*3+j]
        j += 1
    i += 1
    j = 0

print(array1)
```

1. Menentukan hipotesis yang digunakan pada case

*   H0(Hipotesis Null): rata-rata waktu tempuh transportasi kurang dari atau sama dengan 75 menit
*   H1 (Hipotesis Alternatif) : rata-rata waktu tempuh transportasi lebih dari 75 menit

```python
#import library
from scipy import stats
import numpy as np
```

```python
# masukkan data
data = np.array([72, 60, 76, 75, 77, 63, 76, 80, 78, 56, 65, 73, 81, 66,
                 77, 80, 78, 76, 81, 85, 79, 81, 68, 71, 85, 73, 70, 71])
```

```python
# menghitung rata-rata dan standar deviasi
mean = np.mean(data)
std = np.std(data, ddof=1)
```

```python
# menghitung t-score dan p-value
t_score = (mean - 75) / (std / np.sqrt(len(data)))
p_value = 1 - stats.t.cdf(t_score, df=len(data)-1)
print(f't-score: {t_score:.4f}')
print(f'p-value: {p_value:.4f}')
```

```python
# melakukan uji hipotesis
alpha = 0.05
if p_value < alpha:
    print('Reject null hypothesis')
else:
    print('Do not reject null hypothesis')
```

Jika nilai p-value kurang dari tingkat kesalahan alpha, maka kita dapat menolak hipotesis nol dan menerima hipotesis alternatif. Sebaliknya, jika nilai p-value lebih besar atau sama dengan alpha, maka kita tidak dapat menolak hipotesis nol dan tidak memiliki cukup bukti untuk menerima hipotesis alternatif. Pada kasus ini, nilai p-value lebih besar dari alpha, sehingga kita tidak dapat menolak hipotesis nol yang menyatakan bahwa rata-rata waktu tempuh transportasi kurang dari 75 menit. Dengan kata lain, dari data yang ada, tidak ada cukup bukti untuk menyatakan bahwa rata-rata waktu tempuh transportasi melebihi target yang telah ditetapkan oleh CEO perusahaan logistik.
